# How to read an xml file with python

This short notebook shows you how to read the xml file which contains the meta data associated with the REFIT smart home data. This will help you chose and retrieve the *TimeSeriesVariable/@id* for a particular sensor within the homes.

```python
#We can use the xml.etree.ElementTree python module
import xml.etree.ElementTree as ET
```

```python
##Read in the xml file
xml_file = '/Users/PhilSymonds/Documents/workspace/mlsb-sbde/Task1/2070091/REFIT_BUILDING_SURVEY.xml'
tree = ET.parse(xml_file)

#get the root of the directory 
root = tree.getroot()
```

```python
#define an empty dictionary to store the TimeSeriesVariable ids associated with a particular Object (Climate or Building)
variables = {}

#define an empty string variable to store the Climate or Building id
elem_tag = ''

#loop over the branches and sub-braches of the xml file
for elem in root.iter():
    #store the Climate or Building id value
    if ('Climate' in elem.tag) or ('Building' in elem.tag):
        #print(elem.tag, elem.attrib['id'])
        elem_tag = elem.attrib['id']
        #define an empty list to store TimeSeriesVariable ids associated with the Climate/Building
        variables[elem_tag] = []
        
    #find the timeseries id and append values
    if 'TimeSeriesVariable' in elem.tag:
        variables[elem_tag].append(elem.attrib['id'])

##print the timeseries ids and the number or the number of them
print(variables['Climate1'])
print(len(variables['Building01']))
```

```python
#now we can see what these sensors are for (e.g Temperatue, RH, motion, etc...)

#define an empty string to store the zone or space
zone = ''
#loop over brances again
for elem in root.iter():
    #store the current space in the pre-defined zone variable
    if ('Space' in elem.tag):
        zone = elem.attrib['id']
    #print out the variable type associated with a particular TimeSeriesVariable id 
    if elem.attrib['id'] in variables['Building03']:
        print(elem.attrib['id'], zone+' '+elem.attrib['variableType'])
```
